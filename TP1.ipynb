### Arquitectura

Se utiliza una red neuronal de tipo perceptrón multicapa con una capa de entrada de 7 neuronas, dos capas ocultas de 20 neuronas y una capa de salida de una neurona.

Se utilizará el optimizador de Keras *Adam* con una función de error *mean_squared_error* (error cuadrático medio), pasándole batches de 32 registros e iterando 300 veces el modelo para entrenarlo. La función de activación de las neuronas será por *unidad linear rectificada* excepto en la última capa, que será por *función linear (identidad)*

### Carga de datos y codificación de modelo

```python
import os
import sys
# No mostrar WARNINGS
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
from keras import backend as K
from keras.utils import to_categorical

import numpy as np

# Carga de datos en dataframe de pandas
raw = np.genfromtxt('Admission_Predict_Ver1.1.csv', delimiter=",")
raw = np.delete(raw, (0), axis=0)

print("Set de entrenamiento")
raw
```

```python
# Se elimina la columna Serial No. y se divide set de entrenamiento y clases
X = raw[:,1:8]
Y = raw[:,8]

# Se normalizan los valores de las variables independientes
X = X / X.max(axis = 0)

print("Set de entrenamiento normalizado")
X
```

```python
# Definición de arquitectura de la red
learning_rate = 0.001
n_layer_1 = 20
n_layer_2 = 20

def build_regression_model(train_set):
  model = keras.Sequential([
    layers.Dense(train_set.shape[1], activation=tf.nn.relu),
    layers.Dense(n_layer_1, activation=tf.nn.relu),
    layers.Dense(n_layer_2, activation=tf.nn.relu),
    layers.Dense(1, activation='linear')
  ])

  optimizer = tf.keras.optimizers.Adam(learning_rate)

  model.compile(loss='mean_squared_error',
                optimizer=optimizer)
  return model
```

### Set de entrenamiento y set de prueba

```python
# Definición de set de entrenamiento y validación
train_X, test_X, train_Y, test_Y = train_test_split(X, Y, test_size=0.33, random_state=42)

np.set_printoptions(threshold=sys.maxsize)

print("Casos de prueba")
print(train_X)
print("Casos de testeo")
print(test_X)
```

### Entrenamiento

```python
# Construcción de la red
model = build_regression_model(train_X)

# Entrenamiento
history = model.fit(train_X, train_Y, batch_size=32, epochs=300, verbose=1)
```

## Clasificación

### Arquitectura

Se utiliza una red neuronal de tipo perceptrón multicapa con una capa de entrada de 7 neuronas, tres capas ocultas de 68 neuronas y una capa de salida de cinco neuronas (una por cada clase).

Se utilizará el optimizador de Keras Adam con una función de error *categorical_crossentropy* (entropía cruzada categórica), pasándole batches de 32 registros e iterando 300 veces el modelo para entrenarlo. El umbral de activación de las neuronas será por *unidad linear rectificada* excepto en la última capa, que será por *función SoftMax*

### Carga de datos y codificación de modelo

```python
def mapping(x):
  if (x <= 0.5):
    return [1,0,0,0,0];
  if (x <= 0.65):
    return [0,1,0,0,0];
  if (x <= 0.8):
    return  [0,0,1,0,0];
  if (x <= 0.9):
    return  [0,0,0,1,0];
  if (x <= 1):
    return  [0,0,0,0,1];

# Se agrupan los valores de la variable dependiente según requerimiento
Y = np.array(list(map(mapping, Y)))
print(Y)
```

```python
# Definición de arquitectura
def build_classification_model(train_set):
  model = keras.Sequential([
    layers.Dense(68, input_dim=train_set.shape[1], activation=tf.nn.relu),
    layers.Dense(68, activation=tf.nn.relu),
    layers.Dense(5, activation=tf.nn.softmax)
  ])

  optimizer = tf.keras.optimizers.Adam(0.001)

  model.compile(loss='categorical_crossentropy',
                optimizer=optimizer,
                metrics=['accuracy', 'categorical_accuracy', 'mean_squared_error'])
  return model
```

### Set de entrenamiento y set de prueba

```python
# Se separa el set de datos en set de entrenamiento y de prueba
train_X, test_X, train_Y, test_Y = train_test_split(X, Y, test_size=0.33, random_state=42, stratify=Y)

print("Casos de prueba")
print(train_X)
print("Casos de testeo")
print(test_X)
```

### Entrenamiento

```python
# Construcción de la red
model = build_classification_model(train_X)

# Entrenamiento
history = model.fit(train_X, train_Y, batch_size=32, epochs=1000, verbose=1)
```

### Casos de éxito

Se calculan los casos de éxito del modelo entrenado a partir de las predicciones sobre el set de validación

```python
# Predicción
scores = model.evaluate(test_X, test_Y)
print("\nCasos de éxito: %.2f%%" % (scores[1]*100))
```
